```python
import torch
import pandas as pd
```

```python
# üß© Load label encoders
import joblib

label_encoders = joblib.load("data/processed/2025/03/embedding_encoders_march-2025.pkl")
print("‚úÖ Loaded label encoders")

# Optional: Confirm dimensions for sanity check
for key, enc in label_encoders.items():
    print(f"{key}: {len(enc.classes_)} categories")
```

```python
from modeling.transformer_model import RaceTransformer

model = RaceTransformer(
    label_encoders=label_encoders,
    float_dim=63,
    embedding_dim=32,
    nlp_dim=384,
    hidden_dim=128,
    nhead=4,
    num_layers=2
)
```

```python
# 1. Load trained model checkpoint.
model.load_state_dict(torch.load("checkpoints/transformer_2025-04-04T19-08-35/epoch_50.pt"))
model.eval()
```

```python
# 2. Select a race for evaluation.

# Load the training dataset (with winner_flag to allow evaluation of predictions against truth).
df_eval = pd.read_pickle("data/processed/2025/03/model_ready_train.pkl")

# Pick a test race (should still include winner_flag for now).
df_eval = df_eval[df_eval["race_id"] == 891228]  # Example race_id
```

```python
# 3. Batch the race.

# ‚öôÔ∏è Continuous features (numerical signals).
float_cols = [
    "distance_f", "field_size", "class_num", "draw", "age", "or", "rpr", "ts", "lbs",
    "trainer_ovr_runs", "trainer_ovr_wins", "trainer_ovr_win_pct", "trainer_ovr_profit",
    "trainer_last_14_runs", "trainer_last_14_wins", "trainer_last_14_win_pct", "trainer_last_14_profit",
    "jockey_ovr_runs", "jockey_ovr_wins", "jockey_ovr_win_pct", "jockey_ovr_profit",
    "jockey_last_14_runs", "jockey_last_14_wins", "jockey_last_14_win_pct", "jockey_last_14_profit",
    "rpr_rank", "or_rank", "rpr_zscore", "or_zscore"
]

# ‚ûï NLP-derived boolean flags.
nlp_flags = [c for c in df_eval.columns if c.startswith("mentions_")]
float_cols += nlp_flags

# üî¢ Categorical embeddings (integers).
idx_cols = [
    "country_idx", "going_idx", "sex_idx", "type_idx",
    "class_label_idx", "headgear_idx", "race_class_idx", "venue_idx"
]

# üí¨ Text embeddings (vector arrays).
nlp_cols = ["comment_vector", "spotlight_vector"]

print(f"üìä Float features: {len(float_cols)}")
print(f"üî¢ Embedding indices: {idx_cols}")
print(f"üß† NLP fields: {nlp_cols}")

from utils.batching.batch_races import batch_races

eval_batch = batch_races(
    df_eval,
    float_cols=float_cols,
    idx_cols=idx_cols,
    nlp_cols=nlp_cols,
    exclude_non_runners=True
)[0]
```

```python
# 4. Forward pass.
with torch.no_grad():
    logits = model(
        torch.tensor(eval_batch["float_features"], dtype=torch.float32).unsqueeze(0),
        torch.tensor(eval_batch["embedding_indices"], dtype=torch.long).unsqueeze(0),
        torch.tensor(eval_batch["comment_vector"], dtype=torch.float32).unsqueeze(0),
        torch.tensor(eval_batch["spotlight_vector"], dtype=torch.float32).unsqueeze(0),
        torch.tensor(eval_batch["mask"], dtype=torch.bool).unsqueeze(0),
    )
```

```python
# Rank horses.
import torch.nn.functional as F
probs = F.softmax(logits, dim=1).squeeze().numpy()

df_eval = df_eval.copy()
df_eval["prediction_score"] = probs
df_eval = df_eval.sort_values("prediction_score", ascending=False)
df_eval[["name", "prediction_score", "winner_flag", "position"]]
```

```python
# Plot prediction vs true position (scatter).
import matplotlib.pyplot as plt

race_df = df_eval.copy()

# Sort by actual position (1 = winner)
race_df_sorted = race_df.sort_values("position")

plt.figure(figsize=(10, 6))
plt.barh(race_df_sorted["name"], race_df_sorted["prediction_score"], color="skyblue")
plt.gca().invert_yaxis()  # Winner on top
plt.xlabel("Prediction Score")
plt.title("üèá Prediction Scores vs Actual Finishing Order")

# Annotate actual winner
for i, row in race_df_sorted.iterrows():
    if row["winner_flag"] == 1:
        plt.text(row["prediction_score"] + 0.01, i, "üèÜ", fontsize=12, color="red")

# Highlight winner with a different color
colors = ["gold" if p == 1 else "skyblue" for p in race_df["position"]]

plt.barh(race_df["name"], race_df["prediction_score"], color=colors)
plt.xlabel("Prediction Score")
plt.title("üéØ Prediction Scores vs Actual Finishing Order")
plt.tight_layout()
plt.gca().invert_yaxis()
plt.show()
```

```python
# Predicted rank vs true rank scatter plot.
import matplotlib.pyplot as plt

# üß† Create rankings
race_df["predicted_rank"] = race_df["prediction_score"].rank(ascending=False, method="min").astype(int)
race_df["true_rank"] = race_df["position"]

# üéØ Scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(race_df["true_rank"], race_df["predicted_rank"], color="dodgerblue", s=100, edgecolor="k")
plt.plot([1, len(race_df)], [1, len(race_df)], "--", color="gray", label="Perfect Prediction")

# üèá Label each point with horse name
for _, row in race_df.iterrows():
    plt.text(row["true_rank"] + 0.1, row["predicted_rank"], row["name"], fontsize=8)

plt.xlabel("üèÅ True Rank (Finishing Position)")
plt.ylabel("üîÆ Predicted Rank")
plt.title("üéØ Predicted vs True Finish Rank")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
